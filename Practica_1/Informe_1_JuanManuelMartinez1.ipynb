# ¬°RETO!

Escribir una funci√≥n que reciba un diccionario con las notas de los alumnos de un curso y devuelva una serie con la nota m√≠nima, la m√°xima, media y la desviaci√≥n t√≠pica.

```python
def notas_std(di):
    notas = list(di.values())
    nota_max = np.max(notas)
    nota_min = np.min(notas)
    media = np.mean(notas)
    desv_std = np.std(notas)
    
    return([nota_max, nota_min, media, desv_std])
    
```

```python
dic_students = {'Juan': 4.8, 'Andres': 4.5, 'Ana': 5, 'Helena': 2.5, 'Felipe': 1.8, 'Teresa': 4.3, 'David': 3.5}
[max, min, media, desv] = notas_std(dic_students)

print("La nota maxima de la clase es:", max)
print("La nota minima de la clase es:", min)
print("LA nota promedio de la clase es:", np.round(media,3))
print("La desviacion estandar de las notas de la clase es:", np.round(desv,3))
```
