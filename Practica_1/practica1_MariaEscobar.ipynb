```python
import numpy as np #importo la librería Numpy, que usaré para trabajar con arreglos
```

```python
#(a)
#Se crean el par de vectores a y b con la función np.array de Numpy
a=np.array([3.1,1,-0.5,-3.2,6])
b=np.array([1,3,2.2,5.1,1])
```

```python
# (b)
# Implemento multiplicación escalar de a y b
#Se utiliza el producto punto o "dot product" entre los arreglos utilizando comando .dot
escalar=a.dot(b)
print(escalar)
```

```python
#(c)
#Implemento multiplicación punto a punto de los arreglos a y b
c=a*b
print(c)
```

```python
# (d)
#Se construye la matriz A con un arreglo de Numpy
A= np.array([[2,-1,-3],[4,1.5,-2.5],[7.3,-0.9,0.2]])
print(A)
```

```python
# (e)
#Se obtiene la transpuesta de la matriz A con np.transpose
AT=np.transpose(A)
print(AT)
```

```python
# (f)
#np.ones crea una matriz de llena de 1 con las dimenesiones específicadas
unos = np.ones((4, 2))  # Crea una matriz de 4x2 de unos
print(unos)
```

```python
#la función round de Numpy redondea los elementos de una matriz al numero de cifras decimales especificados
numeros = np.array([8.1459, 2.7188, 7.66703])
redondeado = np.round(numeros,2)  # Redondea a 2 decimales
print(redondeado)
```

```python
#La función ceil redondea los numeros de un arreglo llevandolos al numero entero más cercano hacia arriba. Hace referencia a un "techo"
d= np.array([7.7, 0.8, 34.1])
techo = np.ceil(d)  # Redondea hacia arriba
print(techo)
```

```python
#La función floor hace lo contrario a ceil y redondea lo números al siguiente entero hacia abajo
piso=np.floor(d)
print(piso)
```

```python
#(g)
fila= A[0,2] #Accedo a la primera fila, tercera columna de la matriz A
print("Elemento primera fila tercera columna: ")
print(fila)

#(h)
segunda_fila=A[1] #Accedo a la segunda fila
print("Segunda fila matriz A:")
print(segunda_fila)

# (i)
#Utilizo la función np.shape para saber la dimension de la matriz A
print("Dimensión matriz A:")
dim=A.shape
print(dim)
```

```python

import matplotlib.pyplot as plt #primero debo importar la librería matplotlib  que necesitaré para los gráficos
#(j)
# construyo la función y[n] sensoidal con un intervalo específico con Numpy
n = np.arange(0, 101)  # Intervalo de 0 a 100
y = np.sin(np.pi * 0.12 * n)
```

```python
#(k)
y2= np.cos((np.pi*2)*0.03*n)#construcción de otra señal senoidal y2[n]
```

```python
#(l)
#creo señal y3 de la suma de las dos señales y y y2
y3= y+y2

#creo señal y4 que es el producto de las señales y y y2
y4=y*y2
```

```python
#Grafico las señales y[n] y y2[n] con matplotlib
plt.figure(figsize=(10, 6))
plt.plot(n, y, label='y[n]', color='blue')
plt.plot(n, y2, label='y2[n]', color='red')
plt.title('Gráfica de y[n] y y2[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Grafico las señales s[n] llamada y3 en el código y t[n] llamdad y4 con matplotlib
plt.figure(figsize=(10, 6))
plt.plot(n, y3, label='s[n]', color='green')
plt.plot(n, y4, label='t[n]', color='purple')
plt.title('Gráfica de s[n] y t[n]')
plt.xlabel('n')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```
