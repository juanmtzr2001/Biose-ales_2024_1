Representaci√≥n de se√±al impulso unitario generada en clase

```python
def impseq(n0,n1,n2):
    n=np.arange(n1,n2+1) # se crea el vector de muestras
    x=(n-n0)==0

    return [x,n]
```

(1) Representaci√≥n de se√±al escal√≥n unitario.
Se implementa una funci√≥n para una se√±al escal√≥n definida en un intervalo ùëõ1‚â§ùëõ0‚â§ùëõ2

```python
#Se toma el n0 como el punto donde se da un cambio de 0 a 1
def escalon_unitario(n0, n1, n2):

  n=np.arange(n1,n2+1) #vector de muestras con los limites incluidos (eje x)
  x = (n-n0) >=0 # Arroja un True cuando encuentra los valores para ubicar el esacalon
  return [x,n]
```

```python
#Se asigna 7 para n0 y se utiliza la funci√≥n para sacar la se√±al
x,n = escalon_unitario(7, -10, 10)
plt.stem(n,x)
plt.grid()
```

2) Representaci√≥n de se√±al Rampa.
Se crea una funci√≥n para una se√±al rampa definida en un intervalo ùëõ1‚â§ùëõ0‚â§ùëõ2 con una pendiente m.

```python
def rampa(n0, n1, n2,m,t):
  n = np.arange(n1, n2 + 1) # Se crea el vector de muestras con los limites incluidos
  x = (n-n0) >=0
  r=x*((t*n)-n0)*m #Calculo del valor de la se√±al rampa, inicia desde n0
  return [r,n]
```

```python
r,n = rampa(0,-8,8,1,1)
plt.stem(n,r)
plt.grid()
```

Representaci√≥n exponenciales complejas

```python
#n1,n2 = son los l√≠mites inferior y superior de la secuencia
#Sigma = es el coeficiente real de la exponencial
#Omega = es el coeficiente complejo de la exponencial
def exponencial(n1, n2, sigma, omega):
  n = np.arange(n1, n2+1) #Se crea el vector de muestras con los l√≠mites incluidos
  x = np.zeros(n2 - n1 + 1, dtype=complex) #Genero un vector de ceros con la longitud de n, y se ingresa el par√°metro completo
  exp = (sigma + 1j*omega) * n #Calculo del exponencial
  x = np.exp(exp)

  return x
```

```python
ex = exponencial(-8,8,0.2,np.pi)
plt.stem(n,ex) #Se prueba la funci√≥n
plt.grid()
```

Representaci√≥n sinusoides

```python
#n1,n2 = son los limites inferior y superior de la funci√≥n discrtea
#w0 = frecuencia angular
#t0 = fase incial
def sinusoide(n1,n2,w0,t0):
  n = np.arange(n1, n2+1) #Se crea el vector de muestras con los l√≠mites incluidos
  rad=np.deg2rad(w0*n) # Creo variable en la que se guarda frecuencia angular en grados y hace conversi√≥n a radianes
  x = np.sin(rad+t0)

  return [x,n]
```

```python
#prueba de la funci√≥n
s,n0 = sinusoide(-8,8,18,100)
plt.stem(n0,s)
plt.grid()
```

(4) Secuencia X(n)
ùë•[ùëõ]= {ùëü[ùëõ] 0‚â§ùëõ‚â§5ùëü[ùëõ‚àí5] 6‚â§ùëõ‚â§11ùëü[ùëõ‚àí10] 12‚â§ùëõ‚â§17

```python
#Se trata de una secuencia a tramos. Se utiliza la funci√≥n concatenate para unir cada se√±al rampa que representa cada tramo
r1,n1 = rampa(0,0,5,1,1)
r2,n2 = rampa(5,6,11,1,1)
r3,n3 = rampa(10,12,17,1,1)
n=np.concatenate((n1,n2,n3))
x=np.concatenate((r1,r2,r3))
plt.stem(n,x)
```

```python
#se deriva la sencuencia
derivada = np.diff(x)
nn = n[:-1]
plt.stem(nn,derivada)
```

```python
#subplot de la secuencia original y derivada
fig = plt.figure()
ax2 = fig.add_subplot(211)
ax2.stem(nn, derivada)
ax2.set_xlabel('nn')
ax2.set_ylabel('derivada')
ax2.set_title('Secuencia derivada')
ax2.grid()
ax1 = fig.add_subplot(212)
ax1.stem(n, x)
ax1.set_xlabel('n')
ax1.set_ylabel('Eje x')
ax1.set_title('Secuencia original')
ax1.grid()
fig.tight_layout()
plt.show()
```

Se observa que la se√±al derivada de la se√±al rampa original no conserva la misma forma. La se√±al derivada tiene una forma que se asemeja a escal√≥n unitario con entradas de una se√±al impulso con magnitud negativa. Seg√∫n la teor√≠a la derivada de la funci√≥n rampa es la funci√≥n unitaria de Heaviside, y su gr√°fica tiene una pendiente constante en \(x = 0\).
